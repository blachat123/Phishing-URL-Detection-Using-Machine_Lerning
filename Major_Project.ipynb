```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Initialize models again (you might have different variables)
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42)
}

# Dictionary to store ROC info
roc_info = {}

# Re-train all models and store FPR, TPR, AUC
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred_prob = model.predict_proba(X_test)[:,1]  # Probabilities for ROC
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    roc_info[name] = (fpr, tpr, roc_auc)

# Plot all ROC curves
plt.figure(figsize=(10, 8))

for name, (fpr, tpr, roc_auc) in roc_info.items():
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Plot the random guessing line
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison Across Classifiers')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
```

```python
import re
import socket
import urllib
import requests

def extract_features(url):
    features = []

    # Feature 1: Length of URL
    features.append(len(url))

    # Feature 2: Number of dots
    features.append(url.count('.'))

    # Feature 3: Presence of '@' symbol
    features.append(1 if '@' in url else 0)

    # Feature 4: Presence of '-' symbol
    features.append(1 if '-' in url else 0)

    # Feature 5: Number of subdomains
    features.append(url.count('.') - 1)

    # Feature 6: HTTPS token in URL
    features.append(1 if 'https' in url.lower() else 0)

    # Feature 7: IP address usage
    features.append(1 if re.match(r'(\d{1,3}\.){3}\d{1,3}', url) else 0)

    # Feature 8: TinyURL or bit.ly
    features.append(1 if ("tinyurl" in url or "bit.ly" in url) else 0)

    # Feature 9: Length of hostname
    hostname = urllib.parse.urlparse(url).hostname
    features.append(len(hostname) if hostname else 0)

    # Feature 10: Number of digits in URL
    features.append(sum(c.isdigit() for c in url))

    # Feature 11: Number of special characters
    features.append(len(re.findall(r'[^\w]', url)))

    # Feature 12: Is URL using HTTPS scheme
    features.append(1 if url.startswith('https://') else 0)

    # Feature 13: URL contains 'login'
    features.append(1 if 'login' in url.lower() else 0)

    # Feature 14: URL contains 'bank'
    features.append(1 if 'bank' in url.lower() else 0)

    # Feature 15: URL contains 'secure'
    features.append(1 if 'secure' in url.lower() else 0)

    # Feature 16: URL contains 'account'
    features.append(1 if 'account' in url.lower() else 0)

    # Feature 17: URL contains 'update'
    features.append(1 if 'update' in url.lower() else 0)

    # Feature 18: URL contains 'free'
    features.append(1 if 'free' in url.lower() else 0)

    # Feature 19: URL contains 'verify'
    features.append(1 if 'verify' in url.lower() else 0)

    # Feature 20: URL contains 'ebay'
    features.append(1 if 'ebay' in url.lower() else 0)

    # Feature 21: URL contains 'paypal'
    features.append(1 if 'paypal' in url.lower() else 0)

    # Feature 22: URL has double slash redirect
    features.append(1 if '//' in url[url.find('//') + 2:] else 0)

    # Feature 23: Number of "//"
    features.append(url.count('//'))

    # Feature 24: Position of first "//"
    features.append(url.find('//'))

    # Feature 25: Ratio of URL length to hostname length
    features.append(len(url) / len(hostname) if hostname else 0)

    # Feature 26: Is URL from known domain (.com, .net, etc.)
    features.append(1 if any(ext in url.lower() for ext in ['.com', '.net', '.org', '.gov', '.edu']) else 0)

    # Feature 27: Number of query parameters
    query = urllib.parse.urlparse(url).query
    features.append(len(query.split('&')) if query else 0)

    # Feature 28: Length of URL path
    path = urllib.parse.urlparse(url).path
    features.append(len(path))

    # Feature 29: Path depth (number of '/')
    features.append(path.count('/'))

    # Feature 30: Presence of suspicious keywords
    suspicious_keywords = ['confirm', 'account', 'secure', 'webscr', 'login', 'ebayisapi', 'banking', 'update']
    features.append(sum(1 for word in suspicious_keywords if word in url.lower()))

    # Feature 31: Length of first directory
    features.append(len(path.split('/')[1]) if len(path.split('/')) > 1 else 0)

    # Feature 32: Is domain an IP address
    features.append(1 if re.match(r'(\d{1,3}\.){3}\d{1,3}', hostname or '') else 0)

    # Feature 33: Number of hyphens
    features.append(url.count('-'))

    # Feature 34: URL uses '@' to mislead
    features.append(1 if '@' in url else 0)

    # Feature 35: Presence of email ID in URL
    features.append(1 if re.search(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}", url) else 0)

    # Feature 36: Check if shortening service (tinyurl, bit.ly etc.)
    features.append(1 if any(service in url for service in ["tinyurl.com", "bit.ly", "goo.gl", "shorte.st"]) else 0)

    # Feature 37: Ratio of digits to total characters
    features.append(sum(c.isdigit() for c in url) / len(url))

    # Feature 38: Presence of suspicious subdomain
    features.append(1 if "com-" in url.lower() or "net-" in url.lower() else 0)

    # Feature 39: Multiple "//"
    features.append(1 if url.count('//') > 1 else 0)

    # Feature 40: Use of port number
    features.append(1 if ':' in urllib.parse.urlparse(url).netloc else 0)

    # Feature 41: Prefix/suffix separated by '-' in domain
    features.append(1 if '-' in (hostname or '') else 0)

    # Feature 42: URL contains 'server'
    features.append(1 if 'server' in url.lower() else 0)

    # Feature 43: Number of parameters in URL
    features.append(query.count('=') if query else 0)

    # Feature 44: URL uses unusual TLDs
    features.append(1 if any(tld in url for tld in ['.xyz', '.top', '.club', '.info']) else 0)

    # Feature 45: Domain length
    features.append(len(hostname) if hostname else 0)

    # Feature 46: Number of underscores
    features.append(url.count('_'))

    # Feature 47: Is URL suspiciously long
    features.append(1 if len(url) > 75 else 0)

    # Feature 48: Count of '%'
    features.append(url.count('%'))

    return np.array(features).reshape(1, -1)  # Important: shape (1, 48)

```
